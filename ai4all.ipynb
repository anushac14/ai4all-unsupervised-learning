```python
import pandas as pd
import os
```

```python
# Load the dataset
file_path = 'data.csv'
df = pd.read_csv(file_path)
```

```python
# Inspect the dataset
print("Initial Dataset Overview:")
df.head()
```

```python
print("\nDataset Information:")
print(df.info())
```

```python
print("\nMissing Values in Each Column:")
df.isnull().sum()
```

```python
# Remove columns with more than 90% missing values
threshold = 0.9 * len(df)  
df_cleaned = df.dropna(axis=1, thresh=threshold)
```

```python
# check if 'Selling Price' is converted to numeric 

if 'Selling Price' in df_cleaned.columns:
    # Print first few values to inspect issues (optional)
    print(df_cleaned['Selling Price'].head())

    # Ensure all prices are strings and remove any unwanted characters
    df_cleaned['Selling Price'] = (
        df_cleaned['Selling Price']
        .astype(str)
        .str.replace(r'[^\d.]', '', regex=True)  # Remove non-numeric characters
    )

    # Convert the cleaned column to numeric
    df_cleaned['Selling Price'] = pd.to_numeric(df_cleaned['Selling Price'], errors='coerce')

    # Drop rows with NaN values in 'Selling Price'
    df_cleaned = df_cleaned.dropna(subset=['Selling Price'])

# Check the result
df_cleaned.info()
df_cleaned.head()

```

```python
# Fill missing 'Category' values with 'Unknown'
if 'Category' in df_cleaned.columns:
    df_cleaned = df_cleaned.copy() 
    df_cleaned['Category'] = df_cleaned['Category'].fillna('Unknown')
```

```python
# Drop rows with missing 'Selling Price'
df_cleaned.dropna(subset=['Selling Price'], inplace=True)
```

```python
# Handle outliers using IQR
Q1 = df_cleaned['Selling Price'].quantile(0.25)
Q3 = df_cleaned['Selling Price'].quantile(0.75)
IQR = Q3 - Q1
df_cleaned = df_cleaned[(df_cleaned['Selling Price'] >= Q1 - 1.5 * IQR) & 
                        (df_cleaned['Selling Price'] <= Q3 + 1.5 * IQR)]
```

```python
# Scale the 'Selling Price' column
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_cleaned['Selling Price'] = scaler.fit_transform(df_cleaned[['Selling Price']])


df_cleaned['Selling Price'].head()
```

```python
# Drop duplicate rows
df_cleaned.drop_duplicates(inplace=True)
```

```python
# Display summary of the cleaned dataset
print("\nCleaned Dataset Overview:")
df_cleaned.head()
```

```python
print("\nCleaned Dataset Information:")
df_cleaned.info()
```

```python
print("\nRemaining Missing Values in Cleaned Dataset:")
df_cleaned.isnull().sum()
```

```python
# Fill missing text fields with placeholders
df_cleaned['About Product'].fillna('No description', inplace=True)
df_cleaned['Technical Details'].fillna('Not available', inplace=True)

# Verify no missing values remain
print("\nRemaining Missing Values in Cleaned Dataset:")
print(df_cleaned.isnull().sum())
```

```python
# Save the cleaned dataset

df.to_csv("cleaned_amazon_products.csv", index = False)
```
